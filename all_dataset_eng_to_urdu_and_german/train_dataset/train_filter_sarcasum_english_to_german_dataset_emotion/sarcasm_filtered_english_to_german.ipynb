```python
from transformers import pipeline
import pandas as pd
import time
```

```python
translator = pipeline(
    "translation",
    model="facebook/mbart-large-50-many-to-many-mmt",
    device=0
)
```

```python
def translate_into_german(text):
  try:
    if pd.isna(text) or text.strip() == '':
      return ""
    result = translator(str(text))
    return result
  except Exception as e:
    print(f"something wrong {e}")
    return e

```

```python
# translate_german = []
# text_size = len(df)
# batch_size = 100

# for i in range(0 , text_size , batch_size):
#   batch_end = min(i + batch_size , text_size)
#   batch_text = df["text"].iloc[i:batch_end].tolist()
#   print(f"process ...{i+1} ....{batch_end}")

#   batch_german = translator(str(text) for text in batch_text)
#   translate_german.extend(batch_german) #extend -> Ye sirf ek list me doosri list ke sab elements add karta hai (concatenate karta hai).

#   progress = ((i + batch_size ) / text_size) * 100
#   print(f"✅ Progress: {min(progress, 100):.1f}% complete")




translate_german = []
text_size = len(df)
batch_size = 100

for i in range(0, text_size, batch_size):
    batch_end = min(i + batch_size, text_size)
    batch_text = df["text"].iloc[i:batch_end].tolist()
    print(f"process ...{i+1} ....{batch_end}")

    # ✅ Add src_lang and tgt_lang parameters
    batch_german = translator(
        batch_text,  # Remove str() conversion
        src_lang="en_XX",  # English source
        tgt_lang="de_DE"   # German target
    )

    # Extract translation text from results
    translated_texts = [result['translation_text'] for result in batch_german]
    translate_german.extend(translated_texts)

    progress = ((i + batch_size) / text_size) * 100
    print(f"✅ Progress: {min(progress, 100):.1f}% complete")
```

```python
df["translate_german"] = translate_german
file_save = ["binary" , "multiclass" , "translate_german"]
print(f"column save{file_save}")
```
